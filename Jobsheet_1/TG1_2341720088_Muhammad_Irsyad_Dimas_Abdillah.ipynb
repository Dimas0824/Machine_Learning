3. Fungsi masing masing library:
    a. PyPREP adalah porting Python dari PREP pipeline (MATLAB), digunakan untuk membersihkan sinyal EEG dari noise, terutama line noise dan bad channels. PyPREP ini digunakan untuk penelitian neuroengineering atau brain-computer interface (BCI) agar data EEG lebih akurat sebelum dianalisis.

    b. SciPy adalah library fundamental untuk komputasi ilmiah di Python. Berisi modul untuk aljabar linear, optimasi, sinyal, integrasi, statistik, dan lain-lain. Sangat penting untuk pengolahan data, termasuk filtering sinyal ECG/EEG.

    c. wandb adalah alat untuk tracking eksperimen machine learning. Membantu mencatat hasil training (loss, akurasi, hyperparameter) secara real-time, kolaborasi tim, serta membuat dashboard visualisasi eksperimen.

    d. pyECG adalah library python untuk pemrosesan sinyal ECG (Electrocardiogram). Digunakan untuk deteksi puncak R, analisis interval jantung, filtering noise pada sinyal ECG, serta penelitian di bidang kesehatan kardiovaskular.

    d. Reliability and Safety (Keandalan dan Keselamatan):
    Sistem AI yang dipakai untuk diagnosis medis tetapi tidak akurat, sehingga membahayakan pasien.
